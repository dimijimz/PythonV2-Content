```python
seats = 300 #this is the counter. we need this outside of the function to ensure it counts up to this number.
while seats > 0: #here we use the while loop function to tell the computer to keep counting up to 300 which will print ("Sell ticket")
    print("Sell ticket") #we call to print using the new while loop we created
    seats = seats - 1 #we need this stop the condition from holding true so avoid an infinite loop.
```
