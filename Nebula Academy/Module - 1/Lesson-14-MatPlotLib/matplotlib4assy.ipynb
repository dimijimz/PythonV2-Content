Task 1: Load and Examine the Data
Load the CSV File: Load sales_data.csv into a Pandas DataFrame.
Examine the Data: Print the DataFrame and write a brief summary of what the data represents.

```python
# pie_chart_assignment.py

import pandas as pd
import matplotlib.pyplot as plt

# Task 1: Load and Examine the Data
# Load the CSV file
data = pd.read_csv('sales_data.csv')

# Print the DataFrame
print(data)

# Brief summary
# The dataset represents sales figures for different regions. Each row contains a region name and its respective sales value.
```

Task 2: Create a Basic Pie Chart
Generate a Pie Chart: Create a pie chart to visualize sales distribution across regions.
Analysis: Write a brief analysis of what the pie chart shows about the sales distribution.

```python
# Task 2: Create a Basic Pie Chart
# Data for the pie chart
regions = data['Region']
sales = data['Sales']

# Generate a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=regions, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution Across Regions')

# Show the chart
plt.show()

# Brief analysis:
# The pie chart shows the percentage distribution of sales across regions. 
# The region with the largest share is easily identifiable, giving insight into which area performs best in terms of sales.
```

Task 3: Customize the Pie Chart
Customize the Chart:
Add custom colors.
Explode the slice with the highest sales.
Add a shadow to the chart.
Adjust the start angle for better readability.
Save the Chart: Save the customized chart as customized_sales_pie_chart.png.


```python
# Task 3: Customize the Pie Chart

# Identify the index of the highest sales for exploding
explode = [0.1 if s == max(sales) else 0 for s in sales]

# Customize the chart
colors = ['gold', 'lightblue', 'lightgreen', 'pink']  # Custom colors
plt.figure(figsize=(8, 8))
plt.pie(
    sales,
    labels=regions,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    shadow=True,
    colors=colors
)
plt.title('Customized Sales Distribution Across Regions')

# Save the chart
plt.savefig('customized_sales_pie_chart.png')

# Show the chart
plt.show()
```
